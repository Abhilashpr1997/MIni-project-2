```python
def bank_folder():        #here we creating a folder name 'Mini Pro Bank' in c drive for storing informations
                                    
    path1=os.getcwd()     #here i get current working directory and stored to path1 variable
    path2=path1.replace("\\"," ") # replace \\ to white space and stored to path2 variable
    
    path1=path2.split()       #split the path in path2.the path 1= C: user USER
    path2=path1[0]+" "+path1[1]+" "+path1[2]# i have zero idea about others working directory so i choose the path by doing this 
    
    path_location=path2.replace(" ","\\")#white space replaced by double slash \\ and stored in path location variable
    
    location_name=".Mini_Pro_Bank"# creating a folder name mini pro bank
    new_path_location=os.path.join(path_location,location_name) #joint path with folder and creation banking folder in  
    if os.path.exists(new_path_location):# here i check if this path already exist  or not
        
        return new_path_location  #if it is exist return the path location
        
    else:
        os.makedirs(new_path_location) #if not exist create one and return path location
        print("folder creation successful")
    return new_path_location
```

```python
def call_operations(user_info):        #in this fn we allow user to choose an operation 
    
    print("choose Banking Operations: \n1.Deposite. \n2.Withdrawal. \n3.View Balance. \n")
    option=int(input('\nEnter your Choice(1/2/3): '))
    #this option variable store the users choice
    if option == 1:              #if the user choice is 1 provide an option to enter amount
        amount=float(input("Enter Amount: "))# stored the entered amount in a variable called 'amount'
        
        user_info.deposite(amount) #call the fn that created in a class and 
    elif option == 2:                        #if user choose 2
        amount=float(input("Enter Amount: "))  # stored the entered amount in a variable called 'amount'
        user_info.withdrawal(amount)              #call the fn that created in a class and
    elif option == 3:                            #if user choose 2
        user_info.display_info()                 #call the fn that created in a class and if the choice none of them
    else:
        print('Error wrong username or password')  #and print an error message to aware the user 
    return                                           # that you are entered a wrong choice
```

```python
def value_update(account_no,amount):     # in this fn used to update values
    
    path=bank_folder()                   # taking the path from bank folder
            
    rearrange_path=path.replace("\\","/")+"/"+str(account_no)+".txt"   #rearranged by python readable way  
  
    if os.path.exists(rearrange_path):      #if the customer is exist then 
        
        with open(rearrange_path,"r") as file: #read that text file and stored in a variable called content
            
            content= file.read()
            content=content.replace("(","") # the string shows in tuple type so we need to remove brakets in both end
            content=content.replace(")","")   
            content=content.replace(","," ")  #replace coma with whitespace
            content=content.split()    #then split the result text 
            
            content[4]=str(amount)  #update amount 
            
            content=",".join(content) #then joint the text with new amount

            with open(rearrange_path,"w") as file: #then write back tho the location
                file.write(content)

    print("Money Transfer complete")  #show user to transaction completed
```

```python
def login(account_no):           #this function is used to login the customer
    
    path=bank_folder()             # taking the path from bank folder
    
    rearrange_path=path.replace("\\","/")+"/"+str(account_no)+".txt"     #rearranged by python readable way
                
    if os.path.exists(rearrange_path):     #if the customer is exist then
        with open(rearrange_path,"r") as file:       #read that text file and stored in a variable called content
            content= file.read()
            new_content = content.replace(","," ")    # the string shows in tuple type so we need to remove brakets in both end
                                                       #replace coma with whitespace
            new_content=new_content.split()
            password=new_content[5].replace(")","")
            check=password_check(password)            #calling a password checking fn to check the password is correct or not
            if check==1:                               #if correct return 1 else 0
                
                acc   =   int(new_content[0].replace("(",""))
                name  =   new_content[1].replace("_"," ")
                age   =   int(new_content[2])
                address=  new_content[3].replace("_"," ")    
                balance=  float(new_content[4])
                password= int(new_content[5].replace(")",""))
                print('Login sucessfully completed')
                return acc,name,age,address,balance,password
            
            else:
                print("Login failed\n")
                return     
                      
```

```python
def password_check(original_password):   #in this fn used to check the 
    i=0
    while True:
        user_enter_password=input('password: ')  #allow user to enter the password 
        
        
        if original_password!=user_enter_password and i<2: #if original password not equal to entered password
            i=i+1                                           #and no of try not exceed
            print('wrong password try again')               #inform customer that you entered wrong password
            continue
        elif original_password==user_enter_password:       #if the original password is equal to entered password return 1            
            return 1
        
        elif i>=2:                                    # if the no:of failed attempt exceeds 3 times the loop break and return 0
            
            return 0
            break
```
